**Task 1: Exploratory Data Analysis**

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load dataset
df = pd.read_excel('/content/World_development_mesurement.xlsx')
df
```

```python
df.columns
```

```python
# Basic info
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns=[ 'Number of Records'])
```

```python
df.columns
```

```python
# Fill missing values only in numeric columns
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
df[numeric_cols]
```

```python
# Fill missing strings with a placeholder
string_cols = df.select_dtypes(include=['object']).columns
df[string_cols] = df[string_cols].fillna('Unknown')
df[string_cols]
```

**visualizations**

**Histplot**

```python
import matplotlib.pyplot as plt
import seaborn as sns
for col in numeric_cols:
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

**Boxplot**

```python
for col in numeric_cols:
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

**Correlation Matrix**

```python
df[numeric_cols].corr()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[numeric_cols])
X_scaled
```

**Task 2: Model Building & Fitting**

**K-Means Clustering**

```python
from sklearn.cluster import KMeans

# Build and fit KMeans model
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X_scaled)

# Assign cluster labels
df['KMeans_Cluster'] = kmeans.labels_
```

**Hierarchical Clustering**

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Build linkage matrix
linked = linkage(X_scaled, method='ward')

# Plot dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked)
plt.title('Hierarchical Clustering Dendrogram')
plt.show()
```

**DBSCAN**

```python
from sklearn.cluster import DBSCAN

# Build and fit DBSCAN model
dbscan = DBSCAN(eps=1.5, min_samples=5)
dbscan.fit(X_scaled)

# Assign cluster labels
df['DBSCAN_Cluster'] = dbscan.labels_
```

**Task 3: Model Evaluation**

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# KMeans Evaluation
kmeans_labels = kmeans.labels_
kmeans_silhouette = silhouette_score(X_scaled, kmeans_labels)
kmeans_db = davies_bouldin_score(X_scaled, kmeans_labels)
kmeans_ch = calinski_harabasz_score(X_scaled, kmeans_labels)

print("KMeans Silhouette Score:", kmeans_silhouette)
print("KMeans Davies-Bouldin Index:", kmeans_db)
print("KMeans Calinski-Harabasz Index:", kmeans_ch)
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# DBSCAN labels
dbscan_labels = dbscan.labels_

# Check for valid clustering (at least 2 clusters and noise present)
if len(set(dbscan_labels)) > 1 and -1 in dbscan_labels:
    filtered = dbscan_labels != -1
    dbscan_silhouette = silhouette_score(X_scaled[filtered], dbscan_labels[filtered])
    dbscan_db = davies_bouldin_score(X_scaled[filtered], dbscan_labels[filtered])
    dbscan_ch = calinski_harabasz_score(X_scaled[filtered], dbscan_labels[filtered])

    print("DBSCAN Silhouette Score:", dbscan_silhouette)
    print("DBSCAN Davies-Bouldin Index:", dbscan_db)
    print("DBSCAN Calinski-Harabasz Index:", dbscan_ch)
else:
    print("DBSCAN clustering too sparse for evaluation metrics.")
```

```python
from sklearn.cluster import AgglomerativeClustering

# Fit Hierarchical Clustering
hierarchical = AgglomerativeClustering(n_clusters=kmeans.n_clusters)  # Use same k as KMeans for fair comparison
hierarchical_labels = hierarchical.fit_predict(X_scaled)

# Evaluate Hierarchical Clustering
print("Hierarchical Silhouette Score:", silhouette_score(X_scaled, hierarchical_labels))
print("Hierarchical Davies-Bouldin Index:", davies_bouldin_score(X_scaled, hierarchical_labels))
print("Hierarchical Calinski-Harabasz Index:", calinski_harabasz_score(X_scaled, hierarchical_labels))
```

**visualization**

**KMEANS**

```python
from sklearn.decomposition import PCA

# Reduce dimensions for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plot KMeans clusters
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title("KMeans Clusters (PCA)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.show()
```

**DBSCAN**

```python
# Plot DBSCAN clusters
import numpy as np
plt.figure(figsize=(10, 7))
unique_labels = set(dbscan_labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black color for noise points
        col = 'k'

    class_member_mask = (dbscan_labels == k)
    xy = X_pca[class_member_mask]
    plt.scatter(xy[:, 0], xy[:, 1], c=col, marker='o', s=20, label=f'Cluster {k}' if k != -1 else 'Noise')

plt.title('DBSCAN Clusters (PCA)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()
```

**Hierarchial**

```python
from sklearn.cluster import AgglomerativeClustering

model = AgglomerativeClustering(n_clusters=4)
labels = model.fit_predict(X_scaled)
```

```python
# @title
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# Reduce to 2D using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plot clusters
unique_labels = sorted(set(labels))
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

plt.figure(figsize=(10, 6))
for k, col in zip(unique_labels, colors):
    cluster_mask = (labels == k)
    plt.scatter(X_pca[cluster_mask, 0], X_pca[cluster_mask, 1],
                c=[col], label=f"Cluster {k}", edgecolors='k', s=80)

plt.title("Hierarchical Clustering (PCA)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend(title="Clusters", loc="best")
plt.grid(True)
plt.show()
```

**Models comparison**

```python
import pandas as pd
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# Compute hierarchical metrics inline
hier_silhouette = silhouette_score(X_scaled, hierarchical_labels)
hier_db = davies_bouldin_score(X_scaled, hierarchical_labels)
hier_ch = calinski_harabasz_score(X_scaled, hierarchical_labels)

# Compile results
comparison_df = pd.DataFrame({
    'Method': ['KMeans', 'DBSCAN', 'Hierarchical'],
    'Silhouette Score': [
        round(kmeans_silhouette, 3),
        round(dbscan_silhouette, 3) if 'dbscan_silhouette' in locals() else 'N/A',
        round(hier_silhouette, 3)
    ],
    'Davies-Bouldin Index': [
        round(kmeans_db, 3),
        round(dbscan_db, 3) if 'dbscan_db' in locals() else 'N/A',
        round(hier_db, 3)
    ],
    'Calinski-Harabasz Index': [
        round(kmeans_ch, 3),
        round(dbscan_ch, 3) if 'dbscan_ch' in locals() else 'N/A',
        round(hier_ch, 3)
    ]
})

# Display comparison
print("Clustering Method Comparison")
display(comparison_df)
```
